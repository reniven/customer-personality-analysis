```python
# Setting up the environment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime

print("Setup Complete")

warnings.filterwarnings("ignore") # ignore all warnings
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("imakash3011/customer-personality-analysis")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(f"{path}/marketing_campaign.csv", delimiter= '\t')
```

```python
# Verifying the NaN values per column
df.isnull().sum()
```

```python
# Drop rows with missing values to clean the dataset
df.dropna(inplace= True)
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
# Save the cleaned and transformed dataset to a new CSV file
df.to_csv('marketing_campaign_cleaned.csv', index=False)
```
